```python
import google_colab_selenium as gs
import time
driver = gs.Chrome()
```

```python
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
import pandas as pd
```

```python
# Instantiate options
options = Options()

# Add extra options
options.add_argument("--window-size=1920,1080")  # Set the window size
options.add_argument("--disable-infobars")  # Disable the infobars
options.add_argument("--disable-popup-blocking")  # Disable pop-ups
options.add_argument("--ignore-certificate-errors")  # Ignore certificate errors
options.add_argument("--incognito")  # Use Chrome in incognito mode


# Iniciar el navegador
driver = gs.Chrome(options=options)
driver.get('https://rakion.softnyx.com/Ranking/Personal.aspx')
time.sleep(5)

table_headers = [header.text for header in driver.find_elements(By.XPATH, "//*[@id='ranking']/table/thead/tr/th")]
print(table_headers)

```

```python

# Almacenar los datos extraídos
data = []

# Iterar sobre las primeras 10 páginas
for page_num in range(1, 11):
    # Hacer clic en el enlace de la página correspondiente
    driver.execute_script(f"GetListData('{page_num}')")
    time.sleep(5)

    # Encontrar todas las filas en la tabla
    rows = driver.find_elements(By.XPATH, "//*[@id='ranking']/table/tbody/tr")

    # Iterar sobre las filas y extraer datos
    for row in rows:
        # Reintentar la extracción del texto de las celdas
        # para manejar la excepción StaleElementReferenceException
        try:
            row_data = [cell.text for cell in row.find_elements(By.TAG_NAME, "td")]
            data.append(row_data)
        except StaleElementReferenceException:
            pass

# Crear DataFrame
df = pd.DataFrame(data, columns=table_headers)

# Guardar DataFrame en un archivo Excel
df.to_excel('tabla_rakiontop.xlsx', index=False)

# Cerrar el navegador
driver.quit()
```
